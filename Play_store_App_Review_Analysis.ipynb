## **data cleaning and framing**


````python
# convert reviews type as integer type
play_store_df = play_store_df.astype({"Reviews": int})
````

````python
#drop $ sign from price column
play_store_df['Price'] = play_store_df['Price'].str.replace('$','')
````

````python
#convert price column type as float
play_store_df = play_store_df.astype({"Price": float})
````

````python
#in installs column we replace some sign and convert it into integer data type
play_store_df['Installs'] = play_store_df['Installs'].str.replace(',','')
play_store_df['Installs'] = play_store_df['Installs'].str.replace('+','')
play_store_df = play_store_df.astype({"Installs": int})
````

````python
#In size column apps size in mb and kb
play_store_df["Size"].unique()
````

To simplify working with the data, we converted all the sizes of applications in the dataset to megabytes (MB). This helps in maintaining uniformity as the sizes were initially given in both megabytes (MB) and kilobytes (KB).

````python
#After converted
play_store_df["Size"].unique()
````

````python
#data info after cleaning
play_store_df.info()
````

### Variables Description


**Dataset-1:**

This dataset provides information about various apps, including:

App: The app's name and sometimes a brief description.

Category: The app's type, categorized into 33 different groups.

Size: The amount of space the app occupies during installation.

Rating: The average user score, ranging from 1 to 5.

Reviews: The number of user reviews for the app.

Installs: An estimate of how many times the app has been installed.

Type: Whether the app is free or paid.

Price: The installation cost, zero for free apps.

Content rating: The recommended age group for the app.

Genres: The genre or genres to which the app belongs.

Last updated: The date of the latest app update.

Current Ver: The app's current version number.

Android Ver: The Android operating system version the app is compatible with.

This dataset includes:

App: The app's name and sometimes a brief description.

Translated_Review: User reviews translated into English.

Sentiment: The emotional tone of the review, categorized as 'Positive,' 'Negative,' or 'Neutral.'

Sentiment_Polarity: A numerical value indicating the positivity or negativity of the review.

**What is a Sentiment polarity?**


Sentiment polarity is like a mood detector for text. It tells us if the words used are positive, negative, or neutral, helping us understand the overall feelings or opinions expressed by the user about something.

**What is a sentiment Subjectivity ?**



Subjectivity is like a measure of how much someone's personal feelings or opinions show up in a piece of text, as opposed to just giving straightforward facts. When subjectivity is higher, it means the text is more about personal opinions rather than just stating things as they are.
