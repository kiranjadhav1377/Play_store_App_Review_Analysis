### What did you know about your dataset?
The dataset is like a table with 13 columns and 10,841 rows of information.

Some columns, namely 'Reviews,' 'Size,' 'Installs,' and 'Price,' currently contain data types called 'object.'

In the 'Size' column, the sizes are written as words like 'M' for megabytes or 'k' for kilobytes, making them look like text instead of numbers.

The 'Installs' column has numbers written as words, and there are symbols like ',' and '+' in them.

As for the 'Price' column, the prices are written as words starting with the '$' symbol

Within the 'Price' column, values are expressed as strings denoting the price, preceded by the '$' symbol.

## **data cleaning and framing**


````python
# convert reviews type as integer type
play_store_df = play_store_df.astype({"Reviews": int})
````

````python
#drop $ sign from price column
play_store_df['Price'] = play_store_df['Price'].str.replace('$','')
````

````python
#convert price column type as float
play_store_df = play_store_df.astype({"Price": float})
````

````python
#in installs column we replace some sign and convert it into integer data type
play_store_df['Installs'] = play_store_df['Installs'].str.replace(',','')
play_store_df['Installs'] = play_store_df['Installs'].str.replace('+','')
play_store_df = play_store_df.astype({"Installs": int})
````

````python
#In size column apps size in mb and kb
play_store_df["Size"].unique()
````

To simplify working with the data, we converted all the sizes of applications in the dataset to megabytes (MB). This helps in maintaining uniformity as the sizes were initially given in both megabytes (MB) and kilobytes (KB).

````python
#After converted
play_store_df["Size"].unique()
````

````python
#data info after cleaning
play_store_df.info()
````

After changing the data type, now we have some columns in float and int types.

````python
play_store_df.tail()
````

````python
#unique apps
df_df=play_store_df["App"].nunique()
df_df
````

````python
#drop the duplicate apps
right_df=play_store_df.drop_duplicates(subset='App')
right_df.head()
````

````python
#null values in rating-1463
right_df.isnull().sum()
````

Here, we noticed that 1,463 rows have missing values in the 'Rating' column. To address this, we chose to fill in the missing values with the average (mean) rating value across the entire dataset.

````python
#replace the null value
right_df.Rating.fillna(4.19, inplace=True)
````
